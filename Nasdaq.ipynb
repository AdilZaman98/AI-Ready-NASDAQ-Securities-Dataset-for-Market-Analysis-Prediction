# **Outliers from the Dataset:**

```python
# Step 1: Numeric columns select karo
numeric_cols = df.select_dtypes(include=['number']).columns
print("\nNumeric columns:", list(numeric_cols))

# Step 2: Outliers count karo (aapka original code)
print("\n=== OUTLIER DETECTION ===")
outlier_summary = []
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    if outlier_count > 0:
        print(f"Column '{col}': {outlier_count} outliers")
        outlier_summary.append({'Column': col, 'Outliers': outlier_count})

# Step 3: Data clean karo (aapka original code + improvements)
df_clean = df.copy()
print("\n=== CLEANING DATA ===")
for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    initial_rows = len(df_clean)
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    if len(df_clean) < initial_rows:
        print(f"{col}: {initial_rows - len(df_clean)} rows removed")

print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")

# Step 4-5: Scaled Boxplots (0-50 range) - DO GRAPHSCALED
plt.figure(figsize=(16, 12))
first_10_cols = numeric_cols[:10]

plt.subplot(2, 1, 1)
sns.boxplot(data=df[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Original Data: First 10 Columns (Scaled 0-50)")

plt.subplot(2, 1, 2)
sns.boxplot(data=df_clean[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Cleaned Data: First 10 Columns (Scaled 0-50)")

plt.tight_layout()
plt.show()

# Summary table
print("\n=== SUMMARY ===")
summary_df = pd.DataFrame(outlier_summary)
print(summary_df)
print("Data: df (original), df_clean (cleaned)")
```

## **Cell 5 — Feature Engineering & ML Preprocessing**

```python
df_ml = df.copy()

# ── Label encode categorical columns ─────────────────────────────────────────
le_exchange = LabelEncoder()
le_market   = LabelEncoder()
le_status   = LabelEncoder()
df_ml['exchange_enc']    = le_exchange.fit_transform(df_ml['listing_exchange'].astype(str))
df_ml['market_enc']      = le_market.fit_transform(df_ml['market_category'].astype(str))
df_ml['status_enc']      = le_status.fit_transform(df_ml['financial_status'].astype(str))
df_ml['etf_binary']      = (df_ml['etf'] == 'Y').astype(int)
df_ml['is_normal']       = (df_ml['financial_status'] == 'N').astype(int)
df_ml['has_market_cat']  = (df_ml['market_category'] != '').astype(int)
df_ml['is_nasdaq_listed']= (df_ml['listing_exchange'] == 'Q').astype(int)

# ── Define Risk Tier target ───────────────────────────────────────────────────
def assign_risk_tier(row):
    if row['financial_status'] in ['D','E','Q']:
        return 'HIGH_RISK'
    if row['etf'] == 'Y' or row['market_category'] == 'Q':
        return 'LOW_RISK'
    return 'MEDIUM_RISK'

df_ml['risk_tier'] = df_ml.apply(assign_risk_tier, axis=1)

FEATURES = [
    'exchange_enc','market_enc','status_enc','etf_binary',
    'symbol_length','name_word_count','is_normal',
    'has_market_cat','is_nasdaq_listed'
]
TARGET = 'risk_tier'

X = df_ml[FEATURES]
y = df_ml[TARGET]
le_target = LabelEncoder()
y_enc = le_target.fit_transform(y)

# ── Train / Test split ────────────────────────────────────────────────────────
stratify_param = y_enc if len(np.unique(y_enc)) > 1 and len(y_enc) >= 4 else None
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, random_state=42, stratify=stratify_param
)

# ── Scale features ────────────────────────────────────────────────────────────
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f'Train shape: {X_train.shape} | Test shape: {X_test.shape}')
print('Target distribution:')
print(pd.Series(y).value_counts())
```

## **Cell 6 — Train ML Models (RF · XGB · SVM · MLP · LR)**

```python
results = {}

# ── Random Forest ─────────────────────────────────────────────────────────────
rf = RandomForestClassifier(n_estimators=100, max_depth=8,
                            min_samples_split=5, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
results['Random Forest'] = {
    'model':rf, 'preds':rf_pred, 'scaled':False,
    'accuracy':  accuracy_score(y_test, rf_pred),
    'precision': precision_score(y_test, rf_pred, average='macro', zero_division=0),
    'recall':    recall_score(y_test, rf_pred, average='macro', zero_division=0),
    'f1':        f1_score(y_test, rf_pred, average='macro', zero_division=0),
}
print('Random Forest done.')

# ── XGBoost / GradientBoost fallback ─────────────────────────────────────────
if XGB_AVAILABLE:
    xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=6,
                                   learning_rate=0.1, random_state=42,
                                   eval_metric='mlogloss', verbosity=0)
    model_name_xgb = 'XGBoost'
else:
    xgb_model = GradientBoostingClassifier(n_estimators=100, max_depth=5,
                                            learning_rate=0.1, random_state=42)
    model_name_xgb = 'GradBoost'
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
results[model_name_xgb] = {
    'model':xgb_model, 'preds':xgb_pred, 'scaled':False,
    'accuracy':  accuracy_score(y_test, xgb_pred),
    'precision': precision_score(y_test, xgb_pred, average='macro', zero_division=0),
    'recall':    recall_score(y_test, xgb_pred, average='macro', zero_division=0),
    'f1':        f1_score(y_test, xgb_pred, average='macro', zero_division=0),
}
print(f'{model_name_xgb} done.')

# ── SVM ───────────────────────────────────────────────────────────────────────
if X_train_sc.shape[0] >= 2:
    svm = SVC(kernel='rbf', C=10, gamma='scale', probability=True, random_state=42)
    svm.fit(X_train_sc, y_train)
    svm_pred = svm.predict(X_test_sc)
    results['SVM'] = {
        'model':svm, 'preds':svm_pred, 'scaled':True,
        'accuracy':  accuracy_score(y_test, svm_pred),
        'precision': precision_score(y_test, svm_pred, average='macro', zero_division=0),
        'recall':    recall_score(y_test, svm_pred, average='macro', zero_division=0),
        'f1':        f1_score(y_test, svm_pred, average='macro', zero_division=0),
    }
    print('SVM done.')

# ── MLP Neural Network ────────────────────────────────────────────────────────
if X_train_sc.shape[0] >= 10:
    mlp = MLPClassifier(hidden_layer_sizes=(128,64), activation='relu',
                        max_iter=200, random_state=42, early_stopping=True,
                        validation_fraction=0.1, learning_rate_init=0.001)
    mlp.fit(X_train_sc, y_train)
    mlp_pred = mlp.predict(X_test_sc)
    results['Neural Net'] = {
        'model':mlp, 'preds':mlp_pred, 'scaled':True,
        'accuracy':  accuracy_score(y_test, mlp_pred),
        'precision': precision_score(y_test, mlp_pred, average='macro', zero_division=0),
        'recall':    recall_score(y_test, mlp_pred, average='macro', zero_division=0),
        'f1':        f1_score(y_test, mlp_pred, average='macro', zero_division=0),
    }
    print('Neural Net done.')

# ── Logistic Regression ───────────────────────────────────────────────────────
lr = LogisticRegression(max_iter=300, random_state=42, C=1.0)
lr.fit(X_train_sc, y_train)
lr_pred = lr.predict(X_test_sc)
results['Logistic Reg'] = {
    'model':lr, 'preds':lr_pred, 'scaled':True,
    'accuracy':  accuracy_score(y_test, lr_pred),
    'precision': precision_score(y_test, lr_pred, average='macro', zero_division=0),
    'recall':    recall_score(y_test, lr_pred, average='macro', zero_division=0),
    'f1':        f1_score(y_test, lr_pred, average='macro', zero_division=0),
}
print('Logistic Regression done.')

# ── Summary ───────────────────────────────────────────────────────────────────
print('\nMODEL PERFORMANCE SUMMARY')
print('-'*50)
for name, m in results.items():
    print(f'{name:15s}  Acc: {m["accuracy"]:.4f}  F1: {m["f1"]:.4f}')

best_model_name = max(results, key=lambda k: results[k]['accuracy'])
print(f'\nBest model: {best_model_name}')
```

## **Cell 8 — Time Series: Price Simulation & MLP Forecasting**

```python
np.random.seed(42)

SECURITIES = {
    'AAPL': {'start':130, 'drift':0.12, 'vol':0.22},
    'MSFT': {'start':280, 'drift':0.10, 'vol':0.18},
    'NVDA': {'start':250, 'drift':0.35, 'vol':0.42},
    'TSLA': {'start':200, 'drift':-0.05,'vol':0.55},
    'AMZN': {'start':150, 'drift':0.14, 'vol':0.28},
    'SPY':  {'start':420, 'drift':0.08, 'vol':0.14},
}

# ── Generate GBM price series ─────────────────────────────────────────────────
def generate_price(start, drift, vol, n=500):
    dt = 1/252.0
    rets = np.random.normal(drift*dt, vol*np.sqrt(dt), n)
    prices = np.zeros(n)
    prices[0] = start
    for i in range(1, n):
        prices[i] = prices[i-1] * (1 + rets[i])
    idx = pd.date_range(start='2023-01-01', periods=n, freq='B')
    return pd.Series(prices, index=idx)

price_series = {sym: generate_price(**p) for sym, p in SECURITIES.items()}
for sym, s in price_series.items():
    ret = (float(s.iloc[-1])/float(s.iloc[0]) - 1) * 100
    print(f'{sym}: start=${float(s.iloc[0]):.2f}  latest=${float(s.iloc[-1]):.2f}  return={ret:+.1f}%')

# ── MLP sequence forecasting (LSTM-equivalent) ────────────────────────────────
LOOKBACK = 30
FORECAST_DAYS = 90

aapl = price_series['AAPL']
series_sc = MinMaxScaler()
aapl_scaled = series_sc.fit_transform(aapl.values.reshape(-1,1)).flatten()

def make_sequences(arr, lb):
    X, y = [], []
    for i in range(lb, len(arr)):
        X.append(arr[i-lb:i]); y.append(arr[i])
    return np.array(X), np.array(y)

X_seq, y_seq = make_sequences(aapl_scaled, LOOKBACK)
split = int(len(X_seq) * 0.85)
X_tr, X_te = X_seq[:split], X_seq[split:]
y_tr, y_te = y_seq[:split], y_seq[split:]

lstm_model = MLPRegressor(
    hidden_layer_sizes=(256,128,64,32), activation='tanh',
    max_iter=300, random_state=42, early_stopping=True,
    validation_fraction=0.1, learning_rate_init=0.001, tol=1e-5
)
lstm_model.fit(X_tr, y_tr)

y_pred_sc   = lstm_model.predict(X_te)
y_pred_price  = series_sc.inverse_transform(y_pred_sc.reshape(-1,1)).flatten()
y_actual_price= series_sc.inverse_transform(y_te.reshape(-1,1)).flatten()

mae  = np.mean(np.abs(y_actual_price - y_pred_price))
rmse = np.sqrt(np.mean((y_actual_price - y_pred_price)**2))
mape = np.mean(np.abs((y_actual_price - y_pred_price)/y_actual_price)) * 100
print(f'LSTM-equivalent → MAE={mae:.4f}  RMSE={rmse:.4f}  MAPE={mape:.2f}%')

# ── 90-day forward forecast ───────────────────────────────────────────────────
last_seq = aapl_scaled[-LOOKBACK:].copy()
forecast_scaled = []
for _ in range(FORECAST_DAYS):
    p = lstm_model.predict(last_seq.reshape(1,-1))[0]
    forecast_scaled.append(p)
    last_seq = np.append(last_seq[1:], p)

forecast_prices = series_sc.inverse_transform(
    np.array(forecast_scaled).reshape(-1,1)).flatten()
future_dates = pd.date_range(start=aapl.index[-1]+pd.Timedelta(days=1),
                              periods=FORECAST_DAYS, freq='B')
forecast_series = pd.Series(forecast_prices, index=future_dates)

std_dev  = np.std(y_actual_price - y_pred_price)
ci_steps = 1.96 * std_dev * np.sqrt(np.arange(1, FORECAST_DAYS+1))
ci_upper = forecast_prices + ci_steps
ci_lower = np.maximum(forecast_prices - ci_steps, 0)

print(f'Forecast:  30d=${forecast_prices[29]:.2f}  '
      f'60d=${forecast_prices[59]:.2f}  90d=${forecast_prices[89]:.2f}')

# ── Manual ACF ────────────────────────────────────────────────────────────────
def manual_acf(series, nlags=40):
    arr  = np.asarray(series.diff().dropna())
    n    = len(arr)
    mean = arr.mean()
    var  = np.var(arr)
    return np.array(
        [1.0] + [np.sum((arr[:n-l]-mean)*(arr[l:]-mean))/(n*var)
                 for l in range(1, nlags+1)]
    )

acf_vals = manual_acf(aapl, nlags=40)
print('Time series preparation complete!')
```

## **Cell 10 — NLP: Corpus, TF-IDF, Sentiment Classifier, LDA**

```python
np.random.seed(42)

POSITIVE_TEMPLATES = [
    '{} reports record quarterly earnings, stock surges on strong revenue',
    '{} beats analyst estimates with robust revenue growth this quarter',
    '{} announces major AI partnership driving significant investor excitement',
    '{} stock upgraded to Buy by Goldman Sachs, price target raised sharply',
    '{} dividend increased, share buyback program expanded by the board',
    '{} market share gains accelerate in cloud computing segment globally',
    '{} guidance raised for fiscal year amid strong enterprise demand',
    '{} CEO signals very bullish outlook amid broad macro recovery signs',
]
NEGATIVE_TEMPLATES = [
    '{} misses earnings estimates badly, shares tumble in after hours trading',
    '{} faces regulatory scrutiny, SEC investigation formally launched today',
    '{} lays off fifteen percent of workforce amid major restructuring efforts',
    '{} reports wider than expected quarterly loss on declining revenue',
    '{} stock downgraded to Sell by analysts on stretched valuation concerns',
    '{} CEO resignation announcement sparks serious investor concern about strategy',
    '{} margin compression signals significant pricing power erosion ahead',
    '{} litigation risk rises sharply as major class action lawsuit filed',
]
NEUTRAL_TEMPLATES = [
    '{} reports broadly in line quarterly results, shares remain flat',
    '{} maintains full year guidance, management cautious on macro outlook',
    '{} announces strategic review of core business units going forward',
    '{} completes acquisition successfully, integration timeline on track',
    '{} holds annual investor day presenting long term strategic roadmap',
    '{} appoints new CFO from external hire to lead financial transformation',
]

COMPANIES = [
    'Apple','Microsoft','NVIDIA','Amazon','Tesla','Meta','Netflix',
    'Adobe','Zoom','Salesforce','Oracle','Intel','AMD','Qualcomm',
    'Broadcom','Alphabet','PayPal','Visa','Mastercard','JPMorgan'
]

headlines, labels, companies_list = [], [], []
for company in COMPANIES:
    for _ in range(15):
        r = np.random.random()
        if r < 0.55:
            hl = np.random.choice(POSITIVE_TEMPLATES).format(company)
            labels.append('positive')
        elif r < 0.78:
            hl = np.random.choice(NEGATIVE_TEMPLATES).format(company)
            labels.append('negative')
        else:
            hl = np.random.choice(NEUTRAL_TEMPLATES).format(company)
            labels.append('neutral')
        headlines.append(hl)
        companies_list.append(company)

corpus_df = pd.DataFrame({'headline':headlines,'sentiment':labels,'company':companies_list})
print('Corpus size:', len(corpus_df))
print(corpus_df['sentiment'].value_counts())

# ── TF-IDF ───────────────────────────────────────────────────────────────────
tfidf = TfidfVectorizer(max_features=200, stop_words='english',
                        ngram_range=(1,2), min_df=2)
X_tfidf = tfidf.fit_transform(corpus_df['headline'])
feature_names = tfidf.get_feature_names_out()
print('TF-IDF shape:', X_tfidf.shape)

# ── Sentiment Classifier ──────────────────────────────────────────────────────
le_sent = LabelEncoder()
y_sent  = le_sent.fit_transform(corpus_df['sentiment'])
X_tr_s, X_te_s, y_tr_s, y_te_s = train_test_split(
    X_tfidf, y_sent, test_size=0.25, random_state=42, stratify=y_sent)
sent_clf = LogisticRegression(max_iter=500, C=2.0, random_state=42)
sent_clf.fit(X_tr_s, y_tr_s)
sent_pred = sent_clf.predict(X_te_s)
sent_acc  = accuracy_score(y_te_s, sent_pred)
print(f'Sentiment accuracy: {sent_acc:.2%}')
print(classification_report(y_te_s, sent_pred, target_names=le_sent.classes_))

# ── Company sentiment scores ──────────────────────────────────────────────────
def sent_to_score(s):
    return {'positive':1.0,'neutral':0.0,'negative':-1.0}.get(s, 0)
corpus_df['score'] = corpus_df['sentiment'].map(sent_to_score)
company_sent = corpus_df.groupby('company').agg(
    avg_score=('score','mean'), count=('score','count')
).sort_values('avg_score', ascending=False)
print(company_sent)

# ── LDA Topic Modeling ────────────────────────────────────────────────────────
n_topics  = 6
count_vec = CountVectorizer(max_features=150, stop_words='english', min_df=2)
X_count   = count_vec.fit_transform(corpus_df['headline'])
count_features = count_vec.get_feature_names_out()

lda = LatentDirichletAllocation(n_components=n_topics, random_state=42,
                                 max_iter=30, learning_method='online')
lda.fit(X_count)

topic_names_map = {
    0:'Earnings & Revenue', 1:'Market Signals',
    2:'Corporate Events',   3:'Regulatory & Legal',
    4:'AI & Technology',    5:'Strategy & Growth'
}
print('\nLDA Topics:')
for idx, topic in enumerate(lda.components_):
    top_words = [count_features[i] for i in topic.argsort()[-8:][::-1]]
    print(f'  Topic {idx} [{topic_names_map[idx]}]: {", ".join(top_words)}')

# TF-IDF per class
tfidf_by_class = {}
for cls in le_sent.classes_:
    mask = (corpus_df['sentiment'] == cls).values
    tfidf_by_class[cls] = np.array(X_tfidf[mask].mean(axis=0)).flatten()
```
